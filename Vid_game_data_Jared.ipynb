```python
# Load in games dataset and sales dataseet
best_games= pd.read_csv('games.csv')
game_sales= pd.read_csv('video_games_sales.csv')
```

```python
# show the best games data
display(best_games)
```

```python
#clean columns for best games
best_games_c = best_games.drop(['Backlogs','Wishlist','Times Listed','Unnamed: 0', 'Reviews', 'Summary','Genres'],axis=1)
best_games_c.head()
```

```python
# display game sales data
display(game_sales)
```

```python
# clean columns for game sales
game_sales_c = game_sales.drop(['year'], axis=1)
game_sales_c.head()
```

```python
# rename 'name' column to Title
game_sales_c = game_sales_c.rename(columns={'name':'Title'})
game_sales_c.head()
```

```python
#split the name column on / if it starts with Pokemon
mask = game_sales_c['Title'].str.startswith('Pokemon')
game_sales_c.loc[mask, 'Title'] = game_sales_c.loc[mask,'Title'].str.split('/')
game_sales_c = game_sales_c.explode('Title')

game_sales_c.head(30)
```

```python
#merge data into one dataframe on Title
merged_df = pd.merge(best_games_c, game_sales_c, on='Title', how='outer')
```

```python
#drop duplicates
# merged_df = merged_df.drop_duplicates(keep='first')
merged_df = merged_df.drop_duplicates(subset='Title', keep='first')
```

```python
merged_df
```

```python
merged_df.to_csv('merged_data_4.csv', index=False)
```

```python
success_df = merged_df[merged_df['Rating'] >= 4.0]

success_df
```

```python
success_df.to_csv('successful.csv', index=False)
```

```python
# Create a bar chart of global sales by genre
success_df.groupby('genre')['global_sales'].sum().plot(kind='bar')
plt.title('Global Sales by Genre')
plt.xlabel('Genre')
plt.ylabel('Global Sales (in millions)')
plt.show()
```

```python
# Drop rows with "TBD" as a release date
success_df = success_df[success_df['Release Date'] != 'releases on TBD']

# Convert release date to a datetime format
success_df['Release Date'] = pd.to_datetime(success_df['Release Date'], format='%b %d, %Y')

# Create a line chart of global sales by year
success_df.groupby(success_df['Release Date'].dt.year)['global_sales'].sum().plot()
plt.title('Global Sales by Year')
plt.xlabel('Year')
plt.ylabel('Global Sales (in millions)')
plt.show()
```

```python
# Filter by publishers with at least 10 million global sales
publisher_sales = success_df.groupby('publisher')['global_sales'].sum()
publisher_sales = publisher_sales[publisher_sales >= 10]

# Create a stacked bar chart of global sales by publisher and region
publisher_sales.sort_values().plot(kind='barh', color='blue')
plt.title('Total Global Sales by Publisher')
plt.xlabel('Global Sales (in millions)')
plt.ylabel('Publisher')
plt.show()
```

```python
# Create a horizontal bar chart of top 10 games by global sales
success_df.sort_values('global_sales', ascending=False).head(10).plot(x='Title', y='global_sales', kind='barh')
plt.title('Top 10 Games by Global Sales')
plt.xlabel('Global Sales (in millions)')
plt.ylabel('Title')
plt.show()
```

```python
# Aggregate global sales by genre
genre_sales = success_df.groupby('genre')['global_sales'].sum()

# Create a pie chart of global sales by genre
fig, ax = plt.subplots(figsize=(8, 8))
wedges, _, autotexts = ax.pie(genre_sales, wedgeprops=dict(width=0.5), startangle=-40, counterclock=False,
                              autopct='%1.1f%%', textprops=dict(color="w", size=14))
ax.set_title('Global Sales by Genre')

# Add a legend to the chart
ax.legend(genre_sales.index, title='Genres', loc='center left', bbox_to_anchor=(1, 0, 0.5, 1))

# Set the size of the text inside the wedges to be larger
for autotext in autotexts:
    autotext.set_size(16)

# Adjust the padding around the chart so that the labels do not overlap
plt.tight_layout()

# Display the chart
plt.show()
```

```python
# Calculate the total global sales and percentage for each genre
genre_sales = success_df.groupby('genre')['global_sales'].sum()
genre_sales_pct = genre_sales / genre_sales.sum() * 100

# Group bottom 10% categories into an "Other" category
threshold = genre_sales_pct.quantile(0.30)
other_genre_sales = genre_sales_pct[genre_sales_pct < threshold].sum()
other_genres = len(genre_sales_pct[genre_sales_pct < threshold])
genre_sales_pct = genre_sales_pct[genre_sales_pct >= threshold]
genre_sales_pct.loc['Other'] = other_genre_sales
genre_sales_pct.rename(index={genre_sales_pct.index[-1]: 'Other'}, inplace=True)

# Create a pie chart of global sales by genre
fig, ax = plt.subplots(figsize=(8, 8))
wedges, _, autotexts = ax.pie(genre_sales_pct, wedgeprops=dict(width=0.5), startangle=-40, counterclock=False,
                              autopct='%1.1f%%', textprops=dict(color="w", size=14))
ax.set_title('Global Sales by Genre')

# Add a legend to the chart
ax.legend(genre_sales_pct.index, title='Genres', loc='center left', bbox_to_anchor=(1, 0, 0.5, 1))

# Set the size of the text inside the wedges to be larger
for autotext in autotexts:
    autotext.set_size(16)

# Display the chart
plt.show()
```
