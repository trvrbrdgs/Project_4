```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = "https://steamcharts.com/app/1097150"
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
table = soup.find("table", {"class": "common-table"})
```

```python
fg_df = pd.read_html(str(table))[0]
```

```python
fg_df.to_csv("fall_guys.csv", index=False)
fg_df.head()
```

```python
ax =fg_df.plot(x="Month", y="Avg. Players", kind="bar", width=.8, figsize=(10, 6))
ax.set_xticklabels(fg_df['Month'], rotation=45, ha='right')
ax.invert_xaxis()
plt.title('Player Count for Fall Guys')
plt.xlabel('Month', fontsize=5)
plt.ylabel('Avg. Players')
ax.tick_params(axis='both', which='major', labelsize=5)
plt.show()
```

```python
plt.savefig("FallGuysPlayers.png")
plt.show()
```
