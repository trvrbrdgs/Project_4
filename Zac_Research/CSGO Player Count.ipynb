```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = "https://steamcharts.com/app/730"
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
table = soup.find("table", {"class": "common-table"})
```

```python
csgo_df = pd.read_html(str(table))[0]
```

```python
csgo_df.to_csv("csgo_players.csv", index=False)
csgo_df.head()
```

```python
ax =csgo_df.plot(x="Month", y="Avg. Players", kind="bar", width=.8, figsize=(20, 6))
ax.set_xticklabels(csgo_df['Month'], rotation=45, ha='right')
ax.invert_xaxis()
plt.title('Player Count for Counter-Strike: Global Offensive')
plt.xlabel('Month', fontsize=5)
plt.ylabel('Avg. Players')
ax.tick_params(axis='both', which='major', labelsize=5)
plt.show()
```
