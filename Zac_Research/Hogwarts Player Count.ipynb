```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = "https://steamcharts.com/app/990080"
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
table = soup.find("table", {"class": "common-table"})
```

```python
hogwarts_df = pd.read_html(str(table))[0]
```

```python
hogwarts_df.to_csv("hogwarts.csv", index=False)
hogwarts_df.head()
```

```python
ax =hogwarts_df.plot(x="Month", y="Avg. Players", kind="bar", width=.3, figsize=(5, 6))
ax.set_xticklabels(hogwarts_df['Month'], rotation=45, ha='right')
ax.invert_xaxis()
plt.title('Player Count for Hogwarts')
plt.xlabel('Month', fontsize=10)
plt.ylabel('Avg. Players')
ax.tick_params(axis='both', which='major', labelsize=5)
plt.show()
```

```python
plt.savefig("HogwartsPlayers.png")
plt.show()
```
